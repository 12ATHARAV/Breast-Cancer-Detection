```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#loading data from sklearn

breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
# separating features and target

X = data_frame.drop(columns='label', axis=1)    #whenever we eliminate an colum we write axis = 1 else axis = 0
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#standardize the data

#It's also known as Z-score normalization.

#It calculates the mean and standard deviation for each feature.
# Then transforms each data point: (x - mean) / standard_deviation

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
import tensorflow as tf
tf.random.set_seed(3)             #Machine learning algorithms often involve randomness
from tensorflow import keras       #Setting a fixed seed ensures that these random processes produce the same results each time you run your code
```

```python
# layers of NN

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),         # X_train and Y-train are in the form of 2D arrays so we have convert them into 1D array hence uses flatten
    keras.layers.Dense(35, activation='relu'),       #can give any no.
    keras.layers.Dense(2, activation='sigmoid')     #as 0 and 1 output
])
```

```python
#compilation

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',   #'sparse_categorical_crossentropy' is used for multi-class classification problems where the labels are integers(not one-hot encoded).
              metrics=['accuracy'])   #It measures the proportion of correct predictions (correct labels) among the total number of cases examined.
```

one hot encoding

dog - [1 0 0]
cat - [0 1 0]
horse - [0 0 1]

we use = categorical_crossentropy


```python
# training

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=100)

#epochs = how many times your model has to go through the data
```

```python
#visualizing accuracy and loss

plt.plot(history.history['accuracy'])     #2 times history bcz we are taking output from history
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'], loc = 'upper right')
```

```python
#accuracy model for test data
```

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)   #this will compare X_test_std with Y_test

print(accuracy)
```
